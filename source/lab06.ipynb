```python
# 1
n_df = u_data[['movie_id','rating']].groupby(['movie_id']).agg('count')
```

```python
# 2
Pi_df = u_data[['movie_id','rating']].groupby(['movie_id']).agg('sum') / n_df
```

```python
# 3
m_df = u_data[u_data['rating']>=4][['movie_id','rating']].groupby(['movie_id']).agg('count')
```

```python
# 4
p_df = m_df / n_df.astype(float)
```

```python
m_df.shape, n_df.shape, p_df.shape
```

```python
p_df.fillna(0.0, inplace=True)
```

```python
# 5
mu = float(u_data['rating'].sum() / float(u_data['rating'].count()))
k = 10
```

```python
# 6
Pi_fix_df = (u_data[['movie_id','rating']].groupby(['movie_id']).agg('sum').astype('float') + k*mu) / (n_df.astype(float) + k)
```

```python
z = 3.291 #, u'99.90%']
```

```python
phat1 = z**2 / n_df.astype(float)
```

```python
phat2 = z**2 / (2*n_df.astype(float))
```

```python
phat3 = z**2 / (4*n_df.astype(float)**2)
```

```python
pp_df = (p_df / n_df)*(1 - p_df)
```

```python
# 7
wsi_lower = (p_df + phat2 - z*np.sqrt(pp_df + phat3)) / (1 + phat1)
```

```python
wsi_upper = (p_df + phat2 + z*np.sqrt(pp_df + phat3)) / (1 + phat1)
```

```python
n_df1 = n_df.copy()
n_df1['movie_id'] = n_df1.index.values
```

```python
top10_rates = pd.merge(u_item, n_df1, on='movie_id')[['movie_id', 'movie_title','rating']] \
                 .sort_values(['rating'], ascending=False).head(10)['movie_id'].tolist()
```

```python
Pi_df1 = Pi_df.copy()
Pi_df1['movie_id'] = Pi_df1.index.values
```

```python
top10_average = pd.merge(u_item, Pi_df1, on='movie_id')[['movie_id','movie_title','rating']] \
         .sort_values(['rating'], ascending=False).head(10).sort_values('movie_title')['movie_id'].tolist()
```

```python
Pi_fix_df1 = Pi_fix_df.copy()
Pi_fix_df1['movie_id'] = Pi_fix_df1.index.values
```

```python
top10_rating = pd.merge(u_item, Pi_fix_df1, on='movie_id')[['movie_id', 'movie_title','rating']] \
                 .sort_values(['rating'], ascending=False).head(10)['movie_id'].tolist()
```

```python
wsi_lower1 = wsi_lower.copy()
wsi_lower1['movie_id'] = wsi_lower1.index.values
```

```python
top10_lower = pd.merge(u_item, wsi_lower1, on='movie_id')[['movie_id', 'movie_title','rating']] \
                 .sort_values(['rating'], ascending=False).head(10)['movie_id'].tolist()
```

```python
open('lab06s.json', 'w').write(
    json.dumps(
        {
            "top10_rates": top10_rates,
            "top10_average": top10_average,
            "top10_rating": top10_rating,
            "top10_lower": top10_lower
         }
    )
)
```
